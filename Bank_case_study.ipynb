### Customer Class Hierarchy

This section defines a base `Customer` class and three specialized subclasses: `SalariedCustomer`, `NormalCustomer`, and `CorporateCustomer`. Each class includes a constructor (`__init__`) and a `withdraw` method.

```python
class Customer:
    def __init__(self, name, customer_id, initial_balance=0):
        if initial_balance < 0:
            raise ValueError("Initial balance cannot be negative.")
        self.name = name
        self.customer_id = customer_id
        self.balance = initial_balance
        print(f"Customer {self.name} (ID: {self.customer_id}) created with balance: ${self.balance:.2f}")

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive.")
        self.balance += amount
        print(f"Deposited ${amount:.2f}. New balance: ${self.balance:.2f}")

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive.")
        if self.balance < amount:
            print(f"Insufficient funds. Current balance: ${self.balance:.2f}.")
            return False
        self.balance -= amount
        print(f"Withdrew ${amount:.2f}. New balance: ${self.balance:.2f}")
        return True

    def get_balance(self):
        return self.balance

    def __str__(self):
        return f"Customer Name: {self.name}, ID: {self.customer_id}, Balance: ${self.balance:.2f}"
```

#### SalariedCustomer Class

This class represents a customer who receives a regular salary. It inherits from `Customer` and includes a `monthly_salary` attribute. The `withdraw` method can be customized for salaried customers, for example, to allow overdrafts up to a certain limit based on their salary.

```python
class SalariedCustomer(Customer):
    OVERDRAFT_LIMIT_MULTIPLIER = 0.5 # Allow overdraft up to 50% of monthly salary

    def __init__(self, name, customer_id, monthly_salary, initial_balance=0):
        super().__init__(name, customer_id, initial_balance)
        if monthly_salary <= 0:
            raise ValueError("Monthly salary must be positive.")
        self.monthly_salary = monthly_salary
        print(f"Salaried Customer {self.name} created with monthly salary: ${self.monthly_salary:.2f}")

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive.")

        overdraft_limit = self.monthly_salary * self.OVERDRAFT_LIMIT_MULTIPLIER
        available_funds = self.balance + overdraft_limit

        if available_funds < amount:
            print(f"Insufficient funds for Salaried Customer. Current balance: ${self.balance:.2f}. Overdraft limit: ${overdraft_limit:.2f}. Available for withdrawal: ${available_funds:.2f}")
            return False

        self.balance -= amount
        print(f"Salaried Customer withdrew ${amount:.2f}. New balance: ${self.balance:.2f}")
        return True

    def __str__(self):
        return f"Salaried Customer Name: {self.name}, ID: {self.customer_id}, Balance: ${self.balance:.2f}, Monthly Salary: ${self.monthly_salary:.2f}"
```

#### NormalCustomer Class

This class represents a standard customer with no special privileges. It directly inherits the `withdraw` logic from the base `Customer` class.

```python
class NormalCustomer(Customer):
    def __init__(self, name, customer_id, initial_balance=0):
        super().__init__(name, customer_id, initial_balance)
        print(f"Normal Customer {self.name} created.")

    # No special withdraw logic; uses Customer's withdraw method

    def __str__(self):
        return f"Normal Customer Name: {self.name}, ID: {self.customer_id}, Balance: ${self.balance:.2f}"
```

#### CorporateCustomer Class

This class represents a corporate entity. It inherits from `Customer` and includes a `company_name` attribute. The `withdraw` method can be customized for corporate customers, for instance, to have a higher transaction limit or specific approval processes.

```python
class CorporateCustomer(Customer):
    TRANSACTION_LIMIT = 100000 # Example: High transaction limit for corporate accounts

    def __init__(self, name, customer_id, company_name, initial_balance=0):
        super().__init__(name, customer_id, initial_balance)
        self.company_name = company_name
        print(f"Corporate Customer {self.name} for {self.company_name} created.")

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive.")
        if amount > self.TRANSACTION_LIMIT:
            print(f"Withdrawal amount ${amount:.2f} exceeds corporate transaction limit of ${self.TRANSACTION_LIMIT:.2f}.")
            return False

        # Corporate customers might have different overdraft rules or require specific approvals
        # For this example, we'll use the base Customer's balance check
        if self.balance < amount:
            print(f"Insufficient funds for Corporate Customer. Current balance: ${self.balance:.2f}.")
            return False

        self.balance -= amount
        print(f"Corporate Customer withdrew ${amount:.2f}. New balance: ${self.balance:.2f}")
        return True

    def __str__(self):
        return f"Corporate Customer Name: {self.name}, Company: {self.company_name}, ID: {self.customer_id}, Balance: ${self.balance:.2f}"
```

### Demonstration with Multiple Customers (using a collection)

```python
# Define a list of customer details
customer_data = [
    {"type": "Salaried", "name": "David Lee", "id": "SC004", "salary": 6000, "balance": 1500, "withdrawals": [300, 2000, 4000]},
    {"type": "Normal", "name": "Eva Green", "id": "NC005", "balance": 800, "withdrawals": [150, 700]},
    {"type": "Corporate", "name": "Innovate Corp.", "id": "CC006", "company": "Innovate Solutions", "balance": 75000, "withdrawals": [20000, 60000, 120000]}
]

customers = []
print("--- Creating Customers ---")
for data in customer_data:
    if data["type"] == "Salaried":
        customer = SalariedCustomer(data["name"], data["id"], data["salary"], data["balance"])
    elif data["type"] == "Normal":
        customer = NormalCustomer(data["name"], data["id"], data["balance"])
    elif data["type"] == "Corporate":
        customer = CorporateCustomer(data["name"], data["id"], data["company"], data["balance"])
    else:
        raise ValueError(f"Unknown customer type: {data['type']}")
    customers.append(customer)

print("\n--- Initial Balances for All Customers ---")
for customer in customers:
    print(customer)

print("\n--- Performing Withdrawals for All Customers ---")
for i, customer in enumerate(customers):
    print(f"\n--- Withdrawals for {customer.name} ({customer_data[i]['type']}) ---")
    for amount in customer_data[i]["withdrawals"]:
        customer.withdraw(amount)

print("\n--- Final Balances for All Customers ---")
for customer in customers:
    print(customer)
```

### Adding More Customers via File Access and Appending Results

This section demonstrates how to add new customer data by reading from a text file, performing withdrawal operations, and then appending the results back into the same file.

```python
import os

# Define the filename for customer data
customer_filepath = 'new_customers.txt'

# Prepare new customer data in a simple delimited format
new_customer_raw_data = """
Salaried,Frank White,SC007,7000,2000,500,6000
Normal,Grace Lee,NC008,1200,100,1500
"""

# Write the initial new customer data to the file
try:
    with open(customer_filepath, 'w') as f:
        f.write(new_customer_raw_data.strip())
    print(f"Created '{customer_filepath}' with new customer data.")
except IOError as e:
    print(f"Error writing to file {customer_filepath}: {e}")
```

Now, let's read the new customer data from the file, create customer objects, and add them to our existing list of customers. We'll also define their withdrawal amounts.

```python
new_customers_to_process = []
new_customer_withdrawal_data = []

try:
    with open(customer_filepath, 'r') as f:
        for line in f:
            stripped_line = line.strip()
            if not stripped_line:
                continue

            parts = stripped_line.split(',')

            if len(parts) < 3:
                print(f"Skipping malformed line (not enough parts): {stripped_line}")
                continue

            customer_type = parts[0]
            name = parts[1]
            customer_id = parts[2]

            try:
                if customer_type == "Salaried":
                    if len(parts) < 6: # Need at least type, name, id, salary, initial_balance, one withdrawal
                        print(f"Skipping malformed Salaried customer line (not enough data): {stripped_line}")
                        continue
                    monthly_salary = float(parts[3])
                    initial_balance = float(parts[4])
                    customer = SalariedCustomer(name, customer_id, monthly_salary, initial_balance)
                    withdrawals = [float(w) for w in parts[5:]]
                elif customer_type == "Normal":
                    if len(parts) < 4: # Need at least type, name, id, initial_balance
                        print(f"Skipping malformed Normal customer line (not enough data): {stripped_line}")
                        continue
                    initial_balance = float(parts[3])
                    customer = NormalCustomer(name, customer_id, initial_balance)
                    withdrawals = [float(w) for w in parts[4:]]
                else:
                    print(f"Skipping unknown customer type: {customer_type} in line: {stripped_line}")
                    continue

                new_customers_to_process.append(customer)
                new_customer_withdrawal_data.append({"customer": customer, "withdrawals": withdrawals})
                customers.append(customer) # Add to the main customers list
            except ValueError as ve:
                print(f"Error parsing numeric data in line: {stripped_line}. Details: {ve}")
            except Exception as e: # Catch other potential errors during customer object creation
                print(f"An unexpected error occurred while processing line: {stripped_line}. Details: {e}")

except FileNotFoundError:
    print(f"Error: The file '{customer_filepath}' was not found.")
except IOError as e:
    print(f"Error reading from file {customer_filepath}: {e}")

print("--- New Customers Created from File Data ---")
for cust in new_customers_to_process:
    print(cust)
```

Next, we will perform the withdrawals for these new customers and prepare to append the results to the file.

```python
withdrawal_results = []

print("\n--- Performing Withdrawals for New Customers ---")
for item in new_customer_withdrawal_data:
    customer = item["customer"]
    withdrawals = item["withdrawals"]
    result_lines = [f"\n--- Withdrawals for {customer.name} ({customer.__class__.__name__}) ---"]

    for amount in withdrawals:
        initial_balance = customer.balance
        try:
            success = customer.withdraw(amount)
            status = "Success" if success else "Failed (Insufficient Funds/Limit Exceeded)"
            result_lines.append(f"Attempted withdrawal of ${amount:.2f}: {status}. Balance changed from ${initial_balance:.2f} to ${customer.balance:.2f}")
        except ValueError as ve: # Catch ValueError from withdraw if amount is negative or zero
            result_lines.append(f"Attempted withdrawal of ${amount:.2f}: Failed. Reason: {ve}")
        except Exception as e:
            result_lines.append(f"Attempted withdrawal of ${amount:.2f}: Failed due to unexpected error: {e}")

    result_lines.append(f"Final balance for {customer.name}: ${customer.balance:.2f}")
    withdrawal_results.append("\n".join(result_lines))

# Append the withdrawal results to the file
try:
    with open(customer_filepath, 'a') as f:
        f.write("\n\n--- Withdrawal Results for New Customers ---\n")
        for result_block in withdrawal_results:
            f.write(result_block)
            f.write("\n") # Add an extra newline between customer results
    print(f"\nWithdrawal results appended to '{customer_filepath}'.")
except IOError as e:
    print(f"Error appending to file {customer_filepath}: {e}")
```

Finally, let's view the updated content of the `new_customers.txt` file.

```python
# Display the content of the updated file
customer_filepath = 'new_customers.txt' # Corrected file path
try:
    with open(customer_filepath, 'r') as f:
        print(f.read())
except FileNotFoundError:
    print(f"Error: The file '{customer_filepath}' was not found. Please ensure it was created correctly.")
except IOError as e:
    print(f"Error reading from file {customer_filepath}: {e}")
```

### Storing Customer Data in SQLite RDBMS

This section demonstrates how to create a SQLite database, insert customer data, update records, and retrieve data for further processing.

```python
import sqlite3

# Define the database file name
db_filename = 'customer_database.db'

try:
    # Connect to the SQLite database (it will be created if it doesn't exist)
    conn = sqlite3.connect(db_filename)
    cursor = conn.cursor()

    # Drop the table if it already exists (for clean execution in Colab)
    cursor.execute("DROP TABLE IF EXISTS customers")

    # Create the customers table
    cursor.execute("""
        CREATE TABLE customers (
            customer_id TEXT PRIMARY KEY,
            name TEXT NOT NULL,
            customer_type TEXT NOT NULL,
            balance REAL NOT NULL,
            monthly_salary REAL,
            company_name TEXT
        )
    """)
    conn.commit()
    print(f"Database '{db_filename}' and 'customers' table created successfully.")

except sqlite3.Error as e:
    print(f"SQLite error during database setup: {e}")
finally:
    if 'conn' in locals() and conn:
        conn.close()
        print("Database connection closed.")
```

Now, let's insert the `new_customers_to_process` into our newly created SQLite table.

```python
import sqlite3

db_filename = 'customer_database.db'

try:
    conn = sqlite3.connect(db_filename)
    cursor = conn.cursor()

    for customer_obj in new_customers_to_process:
        customer_type = customer_obj.__class__.__name__
        monthly_salary = None
        company_name = None

        if isinstance(customer_obj, SalariedCustomer):
            monthly_salary = customer_obj.monthly_salary
        elif isinstance(customer_obj, CorporateCustomer):
            company_name = customer_obj.company_name

        cursor.execute("""
            INSERT INTO customers (customer_id, name, customer_type, balance, monthly_salary, company_name)
            VALUES (?, ?, ?, ?, ?, ?)
        """, (
            customer_obj.customer_id,
            customer_obj.name,
            customer_type,
            customer_obj.balance,
            monthly_salary,
            company_name
        ))
    conn.commit()
    print("New customers inserted into the database successfully.")

except sqlite3.Error as e:
    print(f"SQLite error during customer insertion: {e}")
finally:
    if 'conn' in locals() and conn:
        conn.close()
        print("Database connection closed.")
```

Next, we will modify the names of these customers in the database by appending "db" to them.

```python
import sqlite3

db_filename = 'customer_database.db'

try:
    conn = sqlite3.connect(db_filename)
    cursor = conn.cursor()

    # Fetch current names and customer_ids to prepare for update
    cursor.execute("SELECT customer_id, name FROM customers")
    customers_to_update = cursor.fetchall()

    for customer_id, current_name in customers_to_update:
        new_name = current_name + "db"
        cursor.execute("UPDATE customers SET name = ? WHERE customer_id = ?", (new_name, customer_id))
    conn.commit()
    print("Customer names updated in the database by appending 'db'.")

except sqlite3.Error as e:
    print(f"SQLite error during name update: {e}")
finally:
    if 'conn' in locals() and conn:
        conn.close()
        print("Database connection closed.")
```

Finally, we will fetch the data from the database and display it after virtually increasing each customer's balance by $3000 (this change is only for display and not committed back to the database in this step).

```python
import sqlite3

db_filename = 'customer_database.db'

try:
    conn = sqlite3.connect(db_filename)
    cursor = conn.cursor()

    cursor.execute("SELECT customer_id, name, customer_type, balance, monthly_salary, company_name FROM customers")
    rows = cursor.fetchall()

    print("\n--- Customers from DB with Balance + $3000 (Display Only) ---")
    for row in rows:
        customer_id, name, customer_type, balance, monthly_salary, company_name = row
        increased_balance = balance + 3000

        output = f"ID: {customer_id}, Name: {name}, Type: {customer_type}, Original Balance: ${balance:.2f}, New Display Balance: ${increased_balance:.2f}"
        if monthly_salary is not None:
            output += f", Monthly Salary: ${monthly_salary:.2f}"
        if company_name is not None:
            output += f", Company: {company_name}"
        print(output)

except sqlite3.Error as e:
    print(f"SQLite error during data fetching: {e}")
finally:
    if 'conn' in locals() and conn:
        conn.close()
        print("Database connection closed.")
```
